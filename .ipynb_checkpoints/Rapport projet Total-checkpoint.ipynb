Présentation du sujet

La prédiction de la production de futurs puits est une problématique importante. Être capable d’anticiper la performance d’un puits va permettre de planifier les investissements à long terme et d’optimiser les stratégies d’acquisition et de vente de terrain. Usuellement, l’industrie pétrolière s’appuie sur une procédure standard, la Decline Curve Analysis (DCA
http://petrowiki.org/Production_forecasting_decline_curve_analysis).
Cette approche générale ne tient pas compte des spécificités de chaque puits. Un groupe pétrolier cherche à aller plus loin, en s’appuyant sur des modèles plus précis et proches de la réalité. Pour cela, un programme de récolte de données permettant de caractériser la physique de chaque puits a été mis en place. Sur plusieurs années, des informations de géolocalisation, production, géologie ont été extraites et nettoyées. L’objectif est désormais d’utiliser au mieux ces données. Plus particulièrement, vous allez être amené à prédire deux indicateurs de performance : les quantités de gaz (GAS360) et de condensat (CUM360) extraits d’un puits, un an après sa mise en service, c’est-à-dire une fois que le système de production est stabilisé. Pour rajouter une dimension probabiliste au problème, il vous est demandé de ne prédire non pas deux valeurs uniques par puits, mais une surface de confiance (bornes hautes et basses pour le gaz ainsi que le condensat). Cette surface devra être la plus petite possible, tout en contenant les vraies valeurs observées.
Description des données
Deux fichiers seront fournis.
1. Un fichier de données d’apprentissage avec 360 observations. La première colonne contient les identifiants des puits et ne servira pas pour la prédiction. Les colonnes D, E, F contiennent des dates de l’historique du puits. Les autres colonnes contiennent des données physiques normalisées. Les deux dernières colonnes contiennent les productions observées de gaz et d’oil GasCum360, OilCum360.
2. Un fichier de données de test ayant exactement la même structure que le fichier d’apprentissage hormis les deux dernières colonnes. Ce deuxième fichier servira à évaluer votre travail avec la métrique décrite ci-dessous.

Donc, pour chaque puits identifié par son API, il faut fournir les bornes min et max de l’estimation du GAS et du OIL. Pour chaque puit 𝑝, on calculera la surface
𝑆(𝑝) = (𝐺𝑎𝑠360_𝑆𝑈𝑃(𝑝) − 𝐺𝑎𝑠360_𝑆𝑈𝑃(𝑝)) ∗ (𝑂𝑖𝑙360_𝑆𝑈𝑃(𝑝) − 𝑂𝑖𝑙360_𝑆𝑈𝑃(𝑝))
Et une variable binaire 𝑖𝑛(𝑝) = 1 si la prédiction 𝐺𝑎𝑠𝐶𝑢𝑚360(𝑝)∈[𝐺𝑎𝑠360_𝐼𝑁𝐹(𝑝),𝐺𝑎𝑠360_𝑆𝑈𝑃(𝑝)] et 𝑂𝑖𝑙𝐶𝑢𝑚360(𝑝)∈[𝑂𝑖𝑙360_𝐼𝑁𝐹(𝑝),𝑂𝑖𝑙360_𝑆𝑈𝑃(𝑝)] et 𝑖𝑛(𝑝) = 0 sinon. 
Il s’agira alors de minimiser la métrique suivante :
𝑀 = 1/𝑛𝑏𝑝𝑢𝑖𝑡𝑠*(Σ(𝑆(𝑝)𝑖𝑛(𝑝)+10(1−𝑖𝑛(𝑝)))

Les modèles enseignés en cours seront utilisés pour prédire les intervalles mentionnés dans la section Métrique d’évaluation. Le langage de programmation sera Matlab (Octave) ou Python (qui ouvre le droit à un bonus). Le meilleur résultat correspondra à la métrique minimum.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


data=pd.read_csv("C:/Users/Amine/Desktop/Amine/Data science/Proj_2018/DataProject.csv", delimiter=";")
data_copy=pd.read_csv("C:/Users/Amine/Desktop/Amine/Data science/Proj_2018/DataProject.csv", delimiter=";")


#################################              Step 1 : DATA COMPLETION                ###e##################################


#How much values do we have to complete?
tracer=data.isnull().sum()
fig = plt.figure(figsize=(15,15))
ax = tracer.plot(kind='barh')
ax.set_xlabel("Nombre de missing values")
ax.set_ylabel("Nom des colonnes")
ax.set_title("Nombre de missing values dans chaque colonne")
plt.show()

```

```python
#Data are normalised

#We can remove some useless data from the dataset
data=data.drop(["API","Date_Drilling","Date_Completion","Date_Production"], axis=1)

#Data have to be converted into float elements: String to float
columns_names=data.columns
for colonne in columns_names:
    colonne_sans_virgule=data[colonne].str.replace(",",".")
    data[colonne]=colonne_sans_virgule.astype('float')
```

```python

##The dataset is almost ready to be treated, we have to complete it before doing our prediction study
##To do so, we decide to use the easy method of K-nearest neighbours
##Let's start by spliting our dataset into training and testing samples

#Identification of the features
null_series=data.isnull().sum()
data_notnull=null_series[null_series==0]
features=data_notnull.index.tolist()
features.pop(0)##features utilisées dans l'algorithme des k plus proches voisins

#Identification of the targets
null_series=data.isnull().sum()
data_isnull=null_series[null_series!=0]
target=data_isnull.index.tolist()
target

#Splitting of the dataset
train_data=data.dropna(axis=0)
data_to_complete=data[data.isnull().any(axis=1)]

   
```

```python
##With our two dataset we tune K-nearest-neigbours algorithm to have the best score
##Concretely we test different number for the n_neihbours paramater 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

#loop for to test different n_neighbours parameter
rmse_values={}#dictionnary containing the score for each parameter n
for n in range(1,15):
    #instanciation du modele
    knn=KNeighborsRegressor(n_neighbors=n, algorithm='brute')
    train_target=train_data[target]
    train_features=train_data[features]
    
    
    #fit the model to data
    knn.fit(train_features, train_target)
    
    #Prediction
    predictions=knn.predict(data_to_complete[features])
    
    #scoring by using mse and missing values replacement
    data_to_complete_copy=data_to_complete.copy()
    j=0#compteur
    L1=[]
    L2=[]
    for k in target:
        c=data_to_complete_copy[k]
        for i in range(len(predictions)):
            if pd.isnull(c.iloc[i])==True:
                data_to_complete_copy.iloc[i][k]=predictions[i][j]
            else: 
                L1.append(c.iloc[i])
                L2.append(predictions[i][j])
        j+=1
    
    mse=mean_squared_error(L1,L2)
    rmse=np.sqrt(mse)
    rmse_values[n]=rmse

#Plot rmse_values, determine the best score for the n_neighnours parameter
fig1 = plt.figure(figsize=(10,5))
x,y=[],[]
for z in range(len(rmse_values)):
    x.append(z+1)
    y.append(rmse_values[z+1]) 
plt.xlabel("n_neighbors")
plt.ylabel("RMSE")
plt.title("Moyenne des erreurs en fonction du nombre n de voisins") 
plt.plot(x,y)
plt.show()
```

```python
##the best parameter is 3 neighbours

#instanciation du modele
knn=KNeighborsRegressor(n_neighbors=3, algorithm='brute')
train_target=train_data[target]
train_features=train_data[features]


#fit the model to data
knn.fit(train_features, train_target)

#Prediction
predictions=knn.predict(data_to_complete[features])

#scoring by using mse and missing values replacement
data_to_complete_copy=data_to_complete.copy()
j=0#compteur
L1=[]
L2=[]
for k in target:
    c=data_to_complete_copy[k]
    for i in range(len(predictions)):
        if pd.isnull(c.iloc[i])==True:
            data_to_complete_copy.iloc[i][k]=predictions[i][j]
        else: 
            L1.append(c.iloc[i])
            L2.append(predictions[i][j])
    j+=1
```

```python
#Let's check if no missing values have been forgot

data_completed=train_data.append(data_to_complete_copy)

tracer=data_completed.isnull().sum()
fig = plt.figure(figsize=(15,15))
ax1 = tracer.plot(kind='barh')
ax1.set_xlabel("Nombre de missing values")
ax1.set_ylabel("Nom des colonnes")
ax1.set_title("Nombre de missing values dans chaque colonne")
plt.show()
```

```python
##Nous allons comparée la méthode de complétion par KNN aux méthodes de complétions par la Moyenne et par la Médiane

def completionMediane(data):
    
   return data.fillna(data.median())
   
def completionMoyenne(data):
    
    return data.fillna(data.mean())
```

```python
##All missing values have been completed we can move on to the next step.

##################################           Step 2 : PREDICTION         ###################################################

```

```python


##Method which computate the metric by giving her the table with GAS_sup, Gas_inf and Oil_sup, Oil_inf
def metric(tableau_final, test_data):
    tableau_final["S(p)"]=(tableau_final["Gas360_SUP"]-tableau_final["Gas360_INF"])*(tableau_final["Oil360_SUP"]-tableau_final["Oil360_INF"])
    
    L=[]
    for i in range(len(tableau_final)):
        ivG=pd.Interval(left=tableau_final.iloc[i]["Gas360_INF"], right=tableau_final.iloc[i]["Gas360_SUP"])
        ivO=pd.Interval(left=tableau_final.iloc[i]["Oil360_INF"], right=tableau_final.iloc[i]["Oil360_SUP"])
    
        if test_data.iloc[i]["GasCum360"] in ivG and test_data.iloc[i]["OilCum360"] in ivO :
            inp=1
        else:
            inp=0
        L.append(inp)
    tableau_final["In(p)"]=L
    tableau_final["M(p)"]=tableau_final["In(p)"]*tableau_final["S(p)"]+10*(1-tableau_final["In(p)"])    
    M=tableau_final["M(p)"].sum()/len(tableau_final)
    return M




##Method that create the tab with the intervals Gas360_SUP, Gas360_INF , Oil360_SUP, Oil360_INF
##In this function we fix the Gas360_SUP, Gas360_INF,Oil360_SUP,Oil360_INF values
##To do that we computate the variance of our prediction and then multiplate this 
##variance by a coefficient to build the interval
def tableauMAJ(u,v,w,x, tableau_final):
        sigma1 = np.sqrt(stats.tvar(tableau_final["GasCum360"]))
        sigma2 = np.sqrt(stats.tvar(tableau_final["OilCum360"]))
        tableau_final["Gas360_SUP"]=tableau_final["GasCum360"]+u*sigma1
        tableau_final["Gas360_INF"]=tableau_final["GasCum360"]-v*sigma1
        tableau_final["Oil360_SUP"]=tableau_final["OilCum360"]+w*sigma2
        tableau_final["Oil360_INF"]=tableau_final["OilCum360"]-x*sigma2




```

```python
##TRAITEMENT


data_completed['API']=data_copy['API']#data complétée par KNN
data['API']=data_copy['API']
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import statistics as stat
import scipy.stats  as stats
import seaborn as sns
from xgboost import  XGBRegressor
from pandas.plotting import scatter_matrix

#Our features and targets
columns_features=['Surf_X', 'Surf_Y', 'Lateral_Length (ft)', 'Depth_TVD_PPLS (ft)',
       'Erosion_PPLS (ft)', 'Pressure_PPLS (PSI)', 'TOC_PPLS (%)',
       'Vcarb_PPLS', 'Vsand_PPLS', 'Vclay_PPLS', 'PR_PPLS', 'YM_PPLS (PSI)',
       'RHOB_PPLS (g/cc)', 'Res_PPLS (Ohmm)', 'GR_PPLS (API)',
       'DT_PPLS (us/ft)', 'DTs_PPLS (us/ft)', 'Temperature (F)',
       'Temp_Anomaly (F)', 'S3Tect_PPLS (PSI)', 'S3_contrast_PPLS (PSI)',
       'Heat_Flow (W/m2)', 'Zone', 'Nbr_Stages', 'Frac_Gradient (PSI/ft)',
       'Proppant_Designed (kg)', 'Proppant_in_Formation (kg)',
       'Avg_Breakdown_Pressure (KPa)', 'Avg_Treating_Pressure (KPa)',
       'Max_Treating_pressure (KPa)', 'Min_Treating_Pressure (KPa)',
       'Avg_Rate_Slurry (bpm)', 'Max_Rate_Slurry (bpm)',
       'Min_Rate_Slurry (bpm)', 'ShutInPressure_Fil (KPa)',
       'ShutInPressure_Initial (KPa)', 'ISIP (KPa)', 'Shot_Density (shots/ft)',
       'Shot_Total', 'Proppant_per_ft (kg/ft)', 'Stage_Spacing (ft)']
columns_target=['GasCum360','OilCum360']

##Predictions 



s1,s2,s3=0,0,0 #counter to make an average of our metrics
for j in range(10):#loop to make the average of the metrics
    
    #First we decided to separate the dataset in two parts, the training part and the testing one

    train_data1, test_data1=train_test_split(data_completed, test_size=0.25)#data completed by KNN method
    train_data2, test_data2=train_test_split(completionMoyenne(data), test_size=0.25)#data comleted by the average
    train_data3, test_data3=train_test_split(completionMediane(data), test_size=0.25)#data completed by the median
    
    #Création des tableaux réponses
    tableau_finalKNN=pd.DataFrame(data=test_data1["API"])
    tableau_finalMoy=pd.DataFrame(data=test_data2["API"])
    tableau_finalMed=pd.DataFrame(data=test_data3["API"])
    
    #computation of the predictions
    for c in columns_target:
        #Knn completion
        X=train_data1[columns_features].as_matrix().astype(np.float64)
        y=train_data1[c].as_matrix().astype(np.float64)
        regressionKNN=svm.SVR()
        regressionKNN.fit(X,y)
        predictions1=regressionKNN.predict(test_data1[columns_features].as_matrix())
    
        tableau_finalKNN[c]=predictions1
        
        #Average completion
        X2=train_data2[columns_features].as_matrix().astype(np.float64)
        y2=train_data2[c].as_matrix().astype(np.float64)
        regressionMoy=svm.SVR()
        regressionMoy.fit(X,y)
        predictions2=regressionMoy.predict(test_data2[columns_features].as_matrix())
        tableau_finalMoy[c]=predictions2
        
        #Median completion
        X3=train_data3[columns_features].as_matrix().astype(np.float64)
        y3=train_data3[c].as_matrix().astype(np.float64)
        regressionMed=svm.SVR()
        regressionMed.fit(X,y)
        predictions3=regressionMed.predict(test_data3[columns_features].as_matrix())
        tableau_finalMed[c]=predictions3
        
    ##We add the predictions in the DataFrame corresponding to the method used   
    tableauMAJ(0.7,0.7,0.5,0.6, tableau_finalKNN)
    tableauMAJ(0.7,0.7,0.5,0.6, tableau_finalMoy)
    tableauMAJ(0.7,0.7,0.5,0.6, tableau_finalMed)
    
    MKNN=metric(tableau_finalKNN,test_data1)
    MMoy=metric(tableau_finalMoy,test_data2)
    MMed=metric(tableau_finalMed,test_data3)
    
    s1+=MKNN/10
    s2+=MMoy/10
    s3+=MMed/10
print("Métric Moyenne sur 100 tirages pour différents types de complétions")
print("KNN:",s1,"    Moyenne:",s2,"     Mediane:",s3)

```

```python
#POUR LA ASUITE NOUS ALLONS ESSAYER D'OPTIMISER CETTE METRIQUE  

##pour cela nous allons jouer sur les hyperparamètres C et gamma du svr.
##En effet le gamma est le coefficient qui se situe dans le kernel

##Puis nous allons trouver les meilleurs coefficients possibles de nos intervallesafin de minimiser la surface

## First Method of optimization of the intervalle in which the intervals are symetric
def learn(tableau_final,metric_seuil,time_max,learning_rate, test_data):
    sigma1 = np.sqrt(stats.tvar(tableau_final["GasCum360"]))
    sigma2 = np.sqrt(stats.tvar(tableau_final["OilCum360"]))
    u,v,x,y=0.5,0.5,0.5,0.5
    tableau_final["Gas360_SUP"]=tableau_final["GasCum360"]+u*sigma1
    tableau_final["Gas360_INF"]=tableau_final["GasCum360"]-v*sigma1
    tableau_final["Oil360_SUP"]=tableau_final["OilCum360"]+x*sigma2
    tableau_final["Oil360_INF"]=tableau_final["OilCum360"]-y*sigma2
    X,Y=[],[]
    while metric(tableau_final, test_data)>metric_seuil:
        u+=learning_rate
        v+=learning_rate
        x+=learning_rate
        y+=learning_rate
        tableau_final["Gas360_SUP"]=tableau_final["GasCum360"]+u*sigma1
        tableau_final["Gas360_INF"]=tableau_final["GasCum360"]-v*sigma1
        tableau_final["Oil360_SUP"]=tableau_final["OilCum360"]+x*sigma2
        tableau_final["Oil360_INF"]=tableau_final["OilCum360"]-y*sigma2
        m=metric(tableau_final, test_data)
        X.append(u)
        Y.append(m)
        if u>1: break
    plt.plot(X,Y)
    plt.show()
    minY=min(Y)
    minX=0
    for j,value in enumerate(Y):
        if value==minY: minX=X[j]
    tableau_final["Gas360_SUP"]=tableau_final["GasCum360"]+minX*sigma1
    tableau_final["Gas360_INF"]=tableau_final["GasCum360"]-minX*sigma1
    tableau_final["Oil360_SUP"]=tableau_final["OilCum360"]+minX*sigma2
    tableau_final["Oil360_INF"]=tableau_final["OilCum360"]-minX*sigma2
    return(minY,minX)

print("Schéma représentant la métrique en fonction du coefficient faisant la largeur de l'intervalle de confiance")
print(learn(tableau_finalMoy, 1, 300, 0.1, test_data2 ))
print("Ainsi nous avons désormais un ordre d'idée du coefficient qui optimise la métrique")
```

```python


#Désormais afin d'optimiser les intervalles de confiance nous allons créer des intervalles asymétriques
#Second optimization Method where intervals are asymetric 
def optimisationSurface(inf,sup,leraning_rate, tableau_final, test_data):
    U=np.arange(inf,sup,leraning_rate)
    V=np.arange(inf,sup,leraning_rate)
    W=np.arange(inf,sup,leraning_rate)
    X=np.arange(inf,sup,leraning_rate)
    M=[]
    U1=[]
    V1=[]
    W1=[]
    X1=[]
    for u in U:
        for v in V:
            for w in W:
                for x in X:
                    
                    tableauMAJ(u,v,w,x,tableau_final)
                    m=metric(tableau_final, test_data) 
                    M.append(m)
                    U1.append(u)
                    V1.append(v)
                    W1.append(w)
                    X1.append(x)
                    
    ##Création du Dataframe avec les résultats
    d={"Metric":M,"U":U1,"V":V1,"W":W1,"X":X1}
    df=pd.DataFrame(data=d)
    min=df["Metric"].min()
    print(df.iloc[df[df["Metric"]==min].index])
    
    df.plot(x='U',y='Metric')
    df.plot(x='V',y='Metric')
    df.plot(x='W',y='Metric')
    df.plot(x='X',y='Metric')
    plt.show()
    return df["Metric"].min()


print(optimisationSurface(0.5,1.0,0.1, tableau_finalMoy,test_data2))
```
